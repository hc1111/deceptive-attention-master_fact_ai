```python
import pandas as pd
import os
import numpy as np
logfolders=[x for x in os.listdir() if '-' in x]
```

```python
home=os.getcwd()
```

```python
# os.chdir()
logfolders
```

```python
mega_df=pd.DataFrame()

for l in logfolders:
    os.chdir(os.path.join(home,l))
    files = [x for x in os.listdir() if 'results' in x]
    for y2 in files:

        df=pd.read_fwf(y2,header=None)
        y=[x.split("=") for x in df[0]]
        my_dict={}
        for y1 in y:
            my_dict[y1[0]]=y1[1]
        df=pd.Series(my_dict).to_frame().rename(columns={0:'Result'}).T
        df['Dataset']=y2.split("_")[0]
        df['Task']=l.split("-")[0]
        df['Lambda']=l.split("-")[-1].split("_")[-1]
        df['Obj']=l.split("-")[-1].split("_")[1]
        mega_df=mega_df.append(df)
# df=[x.split]
```
