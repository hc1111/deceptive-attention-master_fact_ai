```python
import pandas as pd
```

```python
log=pd.read_fwf('logs_sst_wiki_model=emb-lstm-att+hammer=0.0+seed=1',delimiter=' ')
```

```python
newlog_df=pd.Series([x for x in log.values if 'ID' not in x])
```

```python
newlog_df=newlog_df.apply(lambda x:x[0])
```

```python
import numpy as np
```

```python
stats={}
counter=0
for y in newlog_df.values:

    for t in ['train','dev','test']:
            
            s1=False
            for stat in  ['ce_loss','entropy_loss','hammer_loss','acc','accuracy','loss','base_ratio' ,'attention_ratio' ,'base_emb_norm' ,'attn_emb_norm,base_h_norm' ,'attn_h_norm']:
                
                if stat in y:
                    if t in y:
                        y=y.replace(",","")
                        dataset=t
                        try:
                            iteration= int(y.split('iter', maxsplit=1)[-1]\
                               .split(maxsplit=1)[0][:-1])

                            if s1!=True:
                                try:
                                    statistic1=stat
                                    if '_loss' in stat:
                                        stat1_val=float(y.split('ce_loss', maxsplit=1)[-1]\
                                       .split(maxsplit=1
                                             )[0][1:-1])
                                    else:
                                        stat1_val=y.split(stat, maxsplit=1)[-1]\
                                   .split(maxsplit=1)[-1].split(maxsplit=1)[0].replace("=","")
                                    s1=True
                                    try:
                                        stat1_val=float(stat1_val)
                                    except:
                                        stat1_val=np.nan
                                    stats[counter]={'Iter':iteration,'stat1':statistic1,'stat1_val':stat1_val,'DS':t}

                                except Exception as e:
                                    print(e)

                            else:
                                try:
                                    statistic1=stat
                                    if '_loss' in stat:
                                        stat1_val=float(y.split('ce_loss', maxsplit=1)[-1]\
                                       .split(maxsplit=1
                                             )[0][1:-1])
                                    else:
                                        stat1_val=y.split(stat, maxsplit=1)[-1]\
                                   .split(maxsplit=1)[-1].split(maxsplit=1)[0].replace("=","")
                                    try:
                                        stat1_val=float(stat1_val)
                                    except:
                                        stat1_val=np.nan
                                    stats[counter]={'Iter':iteration,'stat1':statistic1,'stat1_val':stat1_val,'DS':t}

                                except Exception as e:
                                    print(e)
                        except Exception as e:
                                    print(e)


                
            
                        counter=counter+1
```

```python
ss=pd.DataFrame.from_dict(stats).T
mega_df=pd.DataFrame()
for r in ss.Iter.unique():

    temp=ss.loc[ss.Iter==r]
    temp['stat1_val']=temp['stat1_val'].fillna(0)
    tp=pd.pivot_table(temp,index='DS',values='stat1_val',columns='stat1')
    tp['Iter']=r
    mega_df=mega_df.append(tp)
mega_df.sort_values(['Iter','DS'])
```
